```python
import cv2, os, glob
import pandas as pd
from utils import compute_differentiation_efficiency_index
from matplotlib import pyplot as plt
from scipy import stats
```

Compute the differentiation efficiency index for true and fluorescence images for batches CD03-1 ~ CD03-6.

```python
result = []

for true_ctnt_filename, pred_ctnt_filename in zip(
    glob.glob("../pix2pix/datasets/CM/B/test/*.png"), 
    glob.glob("../pix2pix/datasets/CM/B_predicted/*.png")
):
    
    true_ctnt = cv2.imread(true_ctnt_filename, cv2.IMREAD_GRAYSCALE)/255
    pred_ctnt = cv2.imread(pred_ctnt_filename, cv2.IMREAD_GRAYSCALE)/255
    
    true_efficiency_index = compute_differentiation_efficiency_index(true_ctnt)
    pred_efficiency_index = compute_differentiation_efficiency_index(pred_ctnt)
    
    result.append((os.path.split(true_ctnt_filename)[1], true_efficiency_index, pred_efficiency_index))
    
result = pd.DataFrame(result, columns=["filename", "true_efficiency_index", "pred_efficiency_index"])
result.to_csv("True_and_predicted_differentiation.csv")

result
```

```python
# visualization

plt.scatter(result.true_efficiency_index, result.pred_efficiency_index)
plt.xlabel("true differentiation efficiency index")
plt.ylabel("predicted differentiation efficiency index")
```

```python
# compute the Pearson'r
r, P = stats.pearsonr(result.true_efficiency_index, result.pred_efficiency_index)
print("r = ", r)
print("P = ", P)
```
